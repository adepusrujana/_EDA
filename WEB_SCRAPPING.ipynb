```python
seller = []
sellertype =[]
BHKs = []
location = []
Price = []
Price_per_sqft = []
Area_sqft = []
Cons_status = []

for j in range(1,25):
    URL = 'https://www.makaan.com/hyderabad-residential-property/buy-property-in-hyderabad-city?page={}&_=1687157286044'.format(j)
    print(URL)
    page = requests.get(URL)
    soup = BeautifulSoup(page.text)

   # 1. Seller Name
    seller_name = soup.find_all('a',class_='seller-name')
    for j in seller_name:
        a = j.text
        seller.append(a)
      # 2. Seller type
    seller_type = soup.find_all('div',class_='second-line')
    for j in seller_type:
        a = j.text
        sellertype.append(a)

    # 3. BHK
    BHK = soup.find_all('a',class_='typelink')
    for j in BHK:
        a = j.text
        BHKs.append(a)

    # 4. Location
    loc = soup.find_all('a',class_="loclink")
    for j in loc:
        a = j.text
        location.append(a)

    # 5. Total Price
    price = soup.find_all('td',class_="price")
    for j in price:
        a = j.text
        Price.append(a)

    # 6. price per sqft
    price_per_sqft = soup.find_all('td',class_="lbl rate")
    for j in price_per_sqft:
        a = j.text
        b = a.replace('/ sq ft','')
        Price_per_sqft.append(b)

    # 7. Area in Sqft
    Area_in_sqft = soup.find_all('td',class_='size')
    for j in Area_in_sqft:
        a = j.text
        Area_sqft.append(a)

    # 8. Construction status
    cons_stus = soup.find_all('td',class_='val')
    for j in cons_stus :
        a = j.text
        Cons_status.append(a)
```

```python
print(len(seller))
print(len(sellertype))
print(len(BHKs))
print(len(location))
print(len(Price))
print(len(Price_per_sqft))
print(len(Area_sqft))
print(len(Cons_status))
```

```python
# Creating Data Frame

properties_df = pd.DataFrame( {
    'Seller Name' : seller,
    'Seller type' : sellertype,
    'BHK' : BHKs,
    'Location' : location,
    'Total Price' : Price,
    'price per sqft' : Price_per_sqft,
    'Area_sqft' : Area_sqft,
    'Construction status' : Cons_status
})
```

```python
def total_price_conv(x):
    newprice = 0
    if(x.find('L') > 0):
        newprice = float(x.replace('L','').strip())*100000
    elif(x.find('Cr')> 0):
        newprice = float(x.replace('Cr','').strip())*10000000
    return newprice
```

```python
properties_df['Total Price'] = properties_df['Total Price'].apply(total_price_conv)
```

```python
# Converting the dtypes of the each column to actual dtype or required dtype.

properties_df['Total Price'] = properties_df['Total Price'].astype('int64')
properties_df['price per sqft'] = properties_df['price per sqft'].str.replace(',','').astype('int64')
properties_df.Area_sqft = properties_df.Area_sqft.astype('int64')
```
